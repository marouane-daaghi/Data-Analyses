```python
import pandas as pd

years = [str(y) for y in range(2014, 2025)]

 top10 = df_ranked[df_ranked['rank_global'] <= 10].copy()

 morocco = df_ranked[
    df_ranked['Country Name'].str.contains('مغرب|Morocco', case=False, na=False)
].head(1).copy()

 top11 = pd.concat([top10, morocco], ignore_index=True)

top11 = top11.drop_duplicates(subset=['Country Code'], keep='first').reset_index(drop=True)

 top11 = top11.sort_values('rank_global')

 cols = ['rank_global', 'Country Name'] + years + ['avg_growth']
top11 = top11[cols]

top11
```

```python
import matplotlib.pyplot as plt

 top11_plot = top11.sort_values('rank_global')

 countries = top11_plot['Country Name']
growth = top11_plot['avg_growth']

plt.figure(figsize=(10, 6))   

plt.bar(countries, growth)    

plt.xticks(rotation=45, ha='right') 
plt.ylabel('متوسط نمو الناتج المحلي (%)')
plt.title('أعلى 10 اقتصادات نموًا في العالم + المغرب (متوسط 2014–2024)')

plt.tight_layout()
plt.show()
```

```python
selected = df_ranked[df_ranked['rank_global'].isin([1, 3, 123])].copy()
```

```python
mapping = {
    'غيانا': 'Guyana',
    'أيرلندا': 'Ireland',
    'المغرب': 'Morocco'
}

selected.loc[:, 'Country Name'] = selected['Country Name'].map(mapping)
```

```python
import matplotlib.pyplot as plt

selected = selected.sort_values('rank_global')

colors = ['#ADD8E6', '#6495ED', '#00008B']

plt.figure(figsize=(9, 6))

plt.bar(selected['Country Name'], selected['avg_growth'], color=colors)

plt.title('Average Real GDP Growth (2014–2024)', fontsize=16, weight='bold')

plt.ylabel('Growth (%)', fontsize=13)

plt.xticks(fontsize=12)

for i, v in enumerate(selected['avg_growth']):
    plt.text(i, v + 0.3, f"{v:.2f}%", ha='center', fontsize=12, weight='bold')

plt.box(False)

plt.tight_layout()
plt.show()
```
